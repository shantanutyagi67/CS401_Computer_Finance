```python
import numpy as np
import matplotlib.pyplot as plt
from math import log, sqrt, pi, exp
import pandas as pd
from scipy.stats import norm

df = pd.read_excel(r'data.xlsx')
df = df[['Week', 'Stock price($)']]
print(df)

k = 50
sigma = 0.2
r = 0.05
T = 20/52
options = 100000
S = df['Stock price($)'].to_numpy()

def delta(S, K, T, r, sigma):
    return norm.cdf((log(S/K)+(r+0.5*sigma**2)*T)/(sigma*sqrt(T)))

d = np.zeros_like(S)
sp = np.zeros_like(S)
c = np.zeros_like(S)
cu = np.zeros_like(S)
I = np.zeros_like(S)

for i in range(20):
    TT = T - i/52
    d[i] = round(delta(S[i], k, TT, r, sigma),3)
d[i+1] = 1.0

sp[0] = d[0]*options
for i in range(20):
    sp[i+1] = (d[i+1] - d[i])*options
    
for i in range(21):
    c[i] = round(S[i]*sp[i]/1000,1)
    
cu[0] = c[0]
for i in range(20):
    cu[i+1] = round(cu[i]*exp(r/52) + c[i+1],1)
    I[i] = cu[i+1] - cu[i] - c[i+1]
    
plt.grid(True)
plt.xlabel('Week')
plt.ylabel('Delta')
plt.plot(d,alpha=1,linewidth=3)
plt.title('Delta Value for each Week')
plt.show()

df['Delta'] = pd.Series(d, index=df.index)
df['Shares Purchased'] = pd.Series(sp, index=df.index)
df['Cost of Shares($000)'] = pd.Series(c, index=df.index)
df['Cumulative cost with interest($000)'] = pd.Series(cu, index=df.index)
df['Interest($000)'] = pd.Series(I, index=df.index)
df
```
